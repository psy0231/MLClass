```python
# 파일 목록 출력
path = "./train"
file_list = os.listdir(path)

print ("file_list: {}".format(file_list))
```

```python
# 파일명과 확장자의 분리, target(y): 순수 파일명
y_train = []
for i in range(len(file_list)):
    spt = os.path.splitext(file_list[i])
    # print(spt) # ('0', '.jpg')
    y_train.append(int(spt[0])) # 파일명을 값으로 사용   
    
print(y_train)
```
