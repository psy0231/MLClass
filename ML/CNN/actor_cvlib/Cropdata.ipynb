```python
img = cv2.imread('./kim.jpg')      # 김남길

faces, confidence = cvlib.detect_face(img)  # 얼굴 검출
print(faces)      # 얼굴 좌표
print(confidence) # 신뢰도
```

```python
faces_area = []

for face in faces:  # 검출된 이미지가 여러개 인경우 순환
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]

    cv2.rectangle(img, (startX,startY), (endX,endY), (0,255,0), 1)

plt.imshow(img)  # 이미지 출력
plt.axis("off") # 축을 출력하지 말것
plt.show()      # 이미지 출력
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)  # 이미지 출력
plt.axis("off") # 축을 출력하지 말것
plt.show()      # 이미지 출력
```

```python
img = cv2.imread('./kim.jpg')      # 김남길

faces, confidence = cvlib.detect_face(img)  # 얼굴 검출
count = 0

face = faces[0]
(startX,startY) = face[0],face[1]
(endX,endY) = face[2],face[3]
cropped = img[startY-10:endY+10, startX-10:endX+10]

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))  # 이미지 출력
plt.axis("off") # off: 축을 출력하지 말것
plt.show()      # 이미지 출력

# BGR -> RGB 자동 변환되어 이미지로 저장
count += 1
now = datetime.datetime.now()
nowDatetime = now.strftime('%Y%m%d%H%M%S')

cv2.imwrite(nowDatetime + "_" + str(count) + ".jpg", cropped)
```

```python
img = cv2.imread('./lee.jpg')     # 이하늬

faces, confidence = cvlib.detect_face(img)  # 얼굴 검출
count = 0

face = faces[0]
(startX,startY) = face[0],face[1]
(endX,endY) = face[2],face[3]
cropped = img[startY-10:endY+10, startX-10:endX+10]

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))  # 이미지 출력
plt.axis("off") # off: 축을 출력하지 말것
plt.show()      # 이미지 출력

# BGR -> RGB 자동 변환되어 이미지로 저장
count += 1
now = datetime.datetime.now()
nowDatetime = now.strftime('%Y%m%d%H%M%S')

cv2.imwrite(nowDatetime + "_" + str(count) + ".jpg", cropped)
```

```python
img = cv2.imread('./father.jpg')  # 열혈사제 출연진 5명

faces, confidence = cvlib.detect_face(img)  # 얼굴 검출
count = 0

for face in faces:
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]
    cropped = img[startY-10:endY+10, startX-10:endX+10]

    plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))  # 이미지 출력
    plt.axis("off") # off: 축을 출력하지 말것
    plt.show()      # 이미지 출력

    # BGR -> RGB 자동 변환되어 이미지로 저장
    count += 1
    now = datetime.datetime.now()
    nowDatetime = now.strftime('%Y%m%d%H%M%S')

    cv2.imwrite(nowDatetime + "_" + str(count) + ".jpg", cropped)
```
