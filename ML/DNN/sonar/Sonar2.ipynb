### K-fold 교차 검증
- 수집된 데이터(관측치, 행)가 적을 경우 검증 데이터도 적음
- 검증 데이터가 적으면, 검증시 정확도의 차이도 대부분 크게 발생함. 
- 데이터가 대규모로 갈 수도록 어떤 성질을 갖는 모집단의 특성을 잘 반영한 모델이 제작됨.

```python
%autosave 0
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential      # class
from tensorflow.keras.models import load_model      # model 사용
from tensorflow.keras.layers import Dense           # 전결합
from tensorflow.keras.callbacks import EarlyStopping # 학습 자동 중지
from tensorflow.keras import regularizers

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold  # K 겹 교차 검증

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgunsl.ttf").get_name()
rc('font', family=font_name)           # 맑은 고딕 폰트 지정
plt.rcParams["font.size"] = 12         # 글자 크기
# plt.rcParams["figure.figsize"] = (10, 4) # 10:4의 그래프 비율
plt.rcParams['axes.unicode_minus'] = False  # minus 부호는 unicode 적용시 한글이 깨짐으로 설정

# Jupyter에게 matplotlib 그래프를 출력 영역에 표시할 것을 지시하는 명령
%matplotlib inline
```

```python
# 데이터 준비
df = pd.read_csv('./sonar.csv', header=None)
print(type(df))
print(df.head(5))
print(df.info())
```

```python
data = df.values
print(type(data))
print(data.shape)
print(data[0])  # 첫행 데이터
```

```python
# 데이터
# <class 'float'> --> <class 'numpy.float64'>
X = data[:, 0:60].astype(float) 
print(type(X[0][0]))
# print(X[0])
# 모든행의 class만 출력
Y_str = data[:, 60]
print(Y_str.shape)
print(Y_str)
```

```python
# 문자 레이블을 숫자화
# 문자열을 알파벳순으로 그룹화한후 0부터 숫자로 변환
encoder = LabelEncoder()
encoder.fit(Y_str)
# print(Y_str)
Y = encoder.transform(Y_str)
print(Y)
```

```python
seed = 0
# x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2,
#                                                   random_state=seed)
# print(y_val)
# print(y_val.shape)
n_fold = 5  # 100 / 5 = 20%씩 분할하여 검증
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```python
for train, val in skf.split(X, Y): # 5번 순환, 행 index 기준 분할
    print(train.shape)
    print(val.shape)
    print(train) # 행 index ★, 80% 훈련
    print(val)   # 행 index ★, 20% 씩 출력
```

```python
print(val[0:10]) # 검증 행 index
```

```python
print(X[1]) # 행 index 1 직접 지정, 2번째 데이터
```

```python
print(X[[1, 5]]) # row index
```

```python
print(X[1, 5]) # 1행 5열의 데이터(2번째행 6번째열)
```

```python
# 동일한 네트워크 구조에서 모델을 새롭게 생성하여 교차 검증
acc = []  # 5번 검증시의 정확도 저장
count = 0 # 검증 진행 비율

for train, val in skf.split(X, Y): # 행 index 기준 분할
    model = Sequential()

    # model.add(Dense(128, input_dim=60, activation='relu'))
    model.add(Dense(80, input_shape=(60, ), activation='relu',
              kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.002)))
#     model.add(Dense(100, activation='relu',
#               kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.002))) # 입력: 128, 출력: 64 
#     model.add(Dense(60, activation='relu',
#               kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.002))) # 입력: 128, 출력: 64 
    model.add(Dense(30, activation='relu',
              kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.002))) # 입력: 128, 출력: 64 
    model.add(Dense(1, activation='sigmoid')) # 0 ~ 1 사이의 값으로 변경
    model.compile(optimizer='adam', loss='binary_crossentropy', 
                  metrics=['accuracy'])

    # patience=10: overfit 발생
    es = EarlyStopping(monitor='loss', patience=6, restore_best_weights=True)
    # train: 훈련 데이터용 index 80% 적용, val: 검증용 index 20% 적용
    hist = model.fit(X[train], Y[train], 
                     validation_data=(X[val], Y[val]),
                     epochs=1000, 
                     batch_size=5, callbacks=[es])

    test_loss, test_acc = model.evaluate(X[val], Y[val], batch_size=2, 
                                         verbose=0)
    print('▶ 손실값:', test_loss, '/정확도:', (test_acc*100), '%')
    
    acc.append(test_acc) # 검증 정확도 저장
    count = count + 1    # 검증 진행 상항
    print('▶', (count*20), '% 검증됨')
        
    fig, loss_ax = plt.subplots()
    # plt.figure(figsize=(6,4)) # ERROR
    fig.set_size_inches(10, 5)  # 챠트 크기 설정

    acc_ax = loss_ax.twinx()  # 오른쪽 y 출 설정

    # 왼쪽 y 축 설정
    loss_ax.plot(hist.history['loss'], 'y', label='train loss')
    loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')
    loss_ax.set_ylim([0.0, 1.0]) # 값을 반영하여 변경

    # 오른쪽 y 축 설정
    acc_ax.plot(hist.history['accuracy'], 'b', label='train accuracy')
    acc_ax.plot(hist.history['val_accuracy'], 'g', label='val accuracy')
    acc_ax.set_ylim([0.0, 1.0])

    # 축 레이블 설정
    loss_ax.set_xlabel('epoch')  # 학습 횟수
    loss_ax.set_ylabel('loss')   # 오차
    acc_ax.set_ylabel('accuracy') # 정확도

    loss_ax.legend(loc='upper left') # 오차 레이블 위치
    acc_ax.legend(loc='lower left')  # 정확도 레이블 위치

    plt.show()
```

```python
# 검증 정확도 시각화
# x = [1, 2, 3, 4, 5] # 학습 횟수
x = [i for i in range(1, 6)] # 학습 횟수

plt.rcParams["figure.figsize"] = (8, 4) # 8:4의 그래프 비율, subplot 사용시 적용
# plt.figure(figsize=(20, 5))            # subplot 사용시 정상작동안함.
fig = plt.figure()
ax = fig.add_subplot(111) #  fig.add_subplot(1, 1, 1) # 1행 1열 첫번째 plot

plt.plot(x, acc, c='g')
plt.ylim(0, 1)
plt.xlabel('학습 횟수')
plt.ylabel('정확도')

# zip([1, 2, 3, 4, 5], [0.8372093  0.83333333 .....
# (1, 0.8372093), (2, 0.83333333)...
for xy in zip(x, acc): # 같은 행단위 열 결합
    # print(type(xy))  # <class 'tuple'>
    # 출력 문자열, 기준 xy 지점, xytext: 문자열 출력 위치
    ax.annotate('%.2f' % xy[1], xy=xy, xytext=xy)  
plt.grid(True)
plt.show()
```

```python
accuracy = np.array(acc)
# accuracy = accuracy.astype('float')
print(accuracy)
print('모든 데이터를 검증한 평균: {0:.1f}%'.format(np.mean(accuracy) * 100))
```

```python
# 평가(검증) 정확도가 높은이유
# 생성된 모델은 8:2로 데이터 분할이 이루어져있고 
# 평가가 이루어질때 평가 index가 훈련에 속할 경우
# 즉 훈련과 검증이 같은 데이터를 사용하는 index들이 발생함으로 ★
# 정확도가 높아지는 문제가 있음으로 권장하지 않음.
```

```python
# 교차 검증은 훈련시 검증이 동시에 이루어져야하며 훈련 종료후 
# 교차 검증은 동일한 데이터를 분할 하는 것임으로 의미 없음 ★
# 훈련이 종료되면 새로운 데이터를 이용하여 교차 검증을 진행 할 것.
```
